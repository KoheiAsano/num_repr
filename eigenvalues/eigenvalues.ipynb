```python
import numpy as np
A = np.array([[1, 2, 7], [2, 3, 1], [3, 1, 2]])
```

```python
np.linalg.eigvals(A)
```

```python
# Power method
x = np.transpose(np.linalg.eigvals(A))
print(x.size)
print(x[0])
print(np.dot(A,x))
for i in range(100):
    r = np.dot(A,x)[0]/x[0]
    print(r)
    x = np.dot(A,x)
```

```python
# jacobi method
A = np.array([[1, 1/2, -(3**0.5)/2], [1/2, 4, (3**0.5)], [-(3**0.5)/2, (3**0.5), 2]])
print(np.linalg.eigvals(A))
p, q = 1, 2
sheta = -np.pi/6
print(A[p][q], A[q][q], A[p][p], sheta)
U = np.array([[1, 0, 0], [0, np.cos(sheta), np.sin(sheta)], [0, -np.sin(sheta), np.cos(sheta)]])
U_inv = np.linalg.inv(U)
U
```

```python
B = np.dot(np.dot(U_inv, A), U)
B
sheta2 = np.pi/4
U2 = np.array([[np.cos(sheta2), 0, np.sin(sheta2)], [0, 1, 0], [-np.sin(sheta2), 0, np.cos(sheta2)]])
U2
```

```python
U2_inv = np.linalg.inv(U2)
U2_inv
B2 = np.dot(np.dot(U2_inv, B), U2)
B2
```
