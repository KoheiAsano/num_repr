```python
import matplotlib.pyplot as plt
import numpy as np
from sympy import symbols

x = symbols('x')
```

```python
# [a,b)
a,b = -5,5.01

f = x*x*x - 5.0*x*x + 1.01*x + 1.88
```

```python
f.subs(x,0)
```

```python
t = np.arange(a, b, 0.01)
fs = [float(f.subs(x,tt)) for tt in t]
plt.plot(t,fs, 'b')
plt.xlabel('x')
# plt.xlim([-2,2])
plt.ylabel('y')
plt.ylim([-1.2,1.2])
plt.grid()
# plt.xticks([i for i in range(a,int(b))])
plt.show()
```

```python
fd=f.diff(x)
fs = [float(fd.subs(x,tt)) for tt in t]
plt.plot(t,fs, 'b')
plt.xlabel('x')
plt.ylabel('y')

plt.grid()
plt.show()    n
```
